````python
import re
````

````python
def substituir_listas(md_text):
    p1 = r'(?<=\n)([*] .*?)(?=\n\n)'
    p = p1
    
    list_pattern = re.compile(p, flags = re.DOTALL)
    all_lists = list_pattern.finditer(md_text)
    
    for ls in all_lists:
        nova_lista = ls.group()
        nova_lista = r'\begin{itemize}' + '\n' + nova_lista + '\n' + r'\end{itemize}' 
        
        # Substituo os itens principais em cada lista
        item_pattern = re.compile(r'(\n|^)\* ')
        nova_lista = item_pattern.sub(r'\n\\item ', nova_lista)
        
        # Encontro as sublistas
        sublist_pattern = re.compile(r'\t\* .*?(?=\\item|\\end)', flags = re.DOTALL)
        
        match = sublist_pattern.search(nova_lista)
        
        while match != None:
            nova_sublista = match.group()
            subitem_pattern = re.compile(r'\t\* ')
            nova_sublista = subitem_pattern.sub(r'\t\\item ', nova_sublista)
            nova_sublista = '\t' + r'\begin{itemize}' + '\n' + nova_sublista + '\t' + r'\end{itemize}' + '\n'
            start_sub = match.start()
            end_sub = match.end()
            nova_lista = nova_lista[:start_sub] + nova_sublista + nova_lista[end_sub:]
            match = sublist_pattern.search(nova_lista)

        start_list = ls.start()
        end_list = ls.end()
        
        texto_final = md_text[:start_list] + nova_lista + md_text[end_list:]
    
    return texto_final
````

````python
with open('input_ex.md', mode = 'r', encoding = 'utf-8') as md_file:
    md_text = md_file.read()
    tex_text = md_text

    # Substitui H1
    h1_md = re.compile(r'(?:(?<=\n)|^)# (?P<titulo>.*?)\n')
    tex_text = h1_md.sub(r'\\chapter{\g<titulo>}\n', tex_text)

    h2_md = re.compile(r'(?:(?<=\n)|^)## (?P<subtitulo>.*?)\n')
    tex_text = h2_md.sub(r'\\section{\g<subtitulo>}\n', tex_text)    

    disp_eq_pattern = re.compile(r'[$]{2}(.*?)[$]{2}')
    tex_text = disp_eq_pattern.sub(r'\\begin{equation*}\n\1\n\\end{equation*}', tex_text)

    inline_eq_pattern = re.compile(r'([^$]{1})[$]{1}([^$]+?)[$]{1}([^$]{1})')
    tex_text = inline_eq_pattern.sub(r'\1$\2$\3', tex_text)

    bold_it = re.compile(r'(\s|^)\*{3}(.+?)\*{3}')
    tex_text = bold_it.sub(r'\1\\textbf{\\textit{\2}}', tex_text)

    bold = re.compile(r'(\s|^)\*{2}(.+?)\*{2}')
    tex_text = bold.sub(r'\1\\textbf{\2}', tex_text)

    italic = re.compile(r'(\s|^)\*{1}(.+?)\*{1}')
    tex_text = italic.sub(r'\1\\textit{\2}', tex_text)

    quotation_pattern = re.compile(r'(?<=\n)>(.*?)(?=\n\n)', flags = re.DOTALL)
    tex_text = quotation_pattern.sub(r'\\begin{quotation}\n\1\n\\end{quotation}', tex_text)
    
    code = re.compile(r'`{3}(?P<lang>[^\n]*?)\n(?P<code>.+?)`{3}', flags = re.DOTALL)
    tex_text = code.sub(r'\\begin{verbatim}\n\g<code>\\end{verbatim}', tex_text)

    type_write = re.compile(r'(\s|^)[`]{1}(.+?)[`]{1}')
    tex_text = type_write.sub(r'\1\\texttt{\2}', tex_text)

    pic = re.compile(r'<img[ ]+?src[ ]*?=[ ]*?["\'](?P<file>.+?)["\'](?P<tags>.*)>')
    tex_text = pic.sub(r'\\includegraphics[width = .6\\textwidth]{\g<file>}', tex_text)
    # @TODO analisar tags

    bad_chars = re.compile(r'(#|%)')
    tex_text = bad_chars.sub(r'\\\1', tex_text)
    # Problema quando estou digitando códigos
    
    tex_text = substituir_listas(tex_text)
    print(tex_text)
    
````

````python
with open('input_ex.md', mode = 'r', encoding = 'utf-8') as md_file:
    md_text = md_file.read()
    
    # ================== (?:(?<=\n)|^) 
    # "?:" Grupo não indexado
    # "(?<=\n)" o espaço anterior deve ser uma quebra
    # "|^" ou o começo do arquivo
    # "# " Deve conter a hashtag e depois um espaço
    # ==================  (?P<titulo>.*?)\n
    # ?P<titulo> Grupo nomeado com titulo
    # .*? eu quero qualquer caracter quantas vezes necessárias (mas mínima)...
    # \n ... até atingir uma quebra de linha
    h1_md = re.compile(r'(?:(?<=\n)|^)# (?P<titulo>.*?)\n')
    
    Ms = h1_md.finditer(md_text)
    
    for m in Ms:
        print(m.group('titulo'))
````

````python
with open('input_ex.md', mode = 'r', encoding = 'utf-8') as md_file:
    md_text = md_file.read()
    
    # Quantos caracteres necessários entre $$...$$
    disp_eq_pattern = re.compile(r'[$]{2}(.*?)[$]{2}')
    
    Ms = disp_eq_pattern.finditer(md_text)
    
    for m in Ms:
        print(m.group(1))
````

````python
with open('input_ex.md', mode = 'r', encoding = 'utf-8') as md_file:
    md_text = md_file.read()
    
    # [^$]{1} qualquer coisa que não seja um cifrão
    # [$]{1} Seguida por um cifrão
    # ([^$]+?) pelo menos um caracter que nao seja cifrao
    # [$]{1} até o proximo cifrao
    # [^$]{1} sem acabar com cifrao extra
    inline_eq_pattern = re.compile(r'[^$]{1}[$]{1}([^$]+?)[$]{1}[^$]{1}')
    
    Ms = inline_eq_pattern.finditer(md_text)
    
    for m in Ms:
        print(m.group(1))
````

````python
with open('input_ex.md', mode = 'r', encoding = 'utf-8') as md_file:
    md_text = md_file.read()
    
    # <img começa com exatamente isso
    # [ ]+? depois tem um ou mais espacos
    # src esta escrito literalmente isso
    # [ ]*? tem zero ou mais espacos
    # =[ ]*? tem um igual seguido de zero ou mais espacos
    # ["\'] depois vem ou uma aspas dupla ou um aspas simples
    # (?P<file>.+?) grupo nomeado com 1 ou mais caracteres
    # ["\'] acaba com uma aspas dupla ou uma aspas simples
    # (?P<tags>.*) grupo nomeado com zero ou mais caracteres
    # ate acabar com >
    
    pic = re.compile(r'<img[ ]+?src[ ]*?=[ ]*?["\'](?P<file>.+?)["\'](?P<tags>.*)>')
    
    Ms = pic.finditer(md_text)
    
    for m in Ms:
        print(m.group())
        print('Nome do arquivo:', m.group('file'))
        print('Tags:', m.group('tags'), '\n')
````

````python
with open('input_ex.md', mode = 'r', encoding = 'utf-8') as md_file:
    md_text = md_file.read()
    
    code = re.compile(r'`{3}(?P<lang>[^\n]*?)\n(?P<code>.+?)`{3}', flags = re.DOTALL)
    
    Ms = code.finditer(md_text)
    
    for m in Ms:
        if m.group('lang') != '':
            print('Linguagem: ' , m.group('lang'))
        else:
            print('Linguagem: ' , 'SEM LINGUAGEM')
        print('Trecho de código: ', m.group('code'))
        print()
````

### Exemplo Substituindo Negritos, Itálicos

```python
texto = "*Nun64* et sem* eget ***AR+C-U*** rutrum ornare** *tipo de **pardal** et* id **lorem**, ou ainda, *caraca **bixo***. **FIM**"

# Grupo 1: (\s|^) Espaço em branco OU começo de string
# \*{N} numero N de asteriscos
# Grupo 2: (.+?) qualquer caracter com 1 ou mais (de menor tamanho possivel)
# \*{N} numero N de asteriscos
negrito_it = re.compile(r'(\s|^)\*{3}(.+?)\*{3}')
negrito = re.compile(r'(\s|^)\*{2}(.+?)\*{2}')
italico = re.compile(r'(\s|^)\*{1}(.+?)\*{1}')


novo_texto = negrito_it.sub(r'\1<strongemph>\2</strongemph>', texto)
novo_texto = negrito.sub(r'\1<strong>\2</strong>', novo_texto)
novo_texto = italico.sub(r'\1<emph>\2</emph>', novo_texto)

print(novo_texto)
```

### Exemplo IDS de vídeos no youtube

```python
# Context: Will YouTube Ever Run Out Of Video IDs? [https://youtu.be/gocwRvLhDf8]

texto = "https://www.youtube.com/watch?v=K8L6KVGG-7o\nhttps://www.youtube.com/watch?v=YYXdXT2l-Gg&list=PL-osiE80TeTt2d9bfVyTiXJA-UTHn6WwU"

# Quero extrair o ID desses vídeos : K8L6KVGG-7o e YYXdXT2l-Gg
# Os IDS do youtube usam base 64 com ultimos digitos sendo "_" e "-"

# < v= > exatamente isso
# < [\w-]+ > 1 ou mais caracteres, no conjunto "word" (= [a-zA-Z0-9_]) e "-"
padrao_id = re.compile(r'v=([\w-]+)')

padroes_encontrados = padrao_id.finditer(texto)

for padrao_encontrado in padroes_encontrados:
    print(padrao_encontrado.group(1))
```

SAÍDA:

```
K8L6KVGG-7o
YYXdXT2l-Gg
```

### Exemplo encontrando emails válidos

```python
texto = '''bol@uol.com.br
abc.game@email.net
OtherMail123@strange-mail.org
institutional-account@university.edu.br
notaanemail@ola
notanemail2@ponto.com.
'''

# < [\w.-]+ > um conjunto de caracteres word \w (=[a-zA-Z0-9_]), ponto e traço ocorrendo 1 ou mais vezes 
# < @ > obrigatorio
# < [a-z\-]+ > conjunto de caracteres minusculos de a até z ou um traço, ocorrendo 1 ou mais vezes
# < \.(com|edu|net|org) > um ponto obrigatorio seguido de com OU edu OU net OU org
# (\s|\.[a-z]{2}) um espaco vazio OU um ponto seguido de 2 letras minusculas
padrao_email = re.compile(r'[\w.-]+@[a-z\-]+\.(com|edu|net|org)(\s|\.[a-z]{2})')

padroes_encontrados = padrao_email.finditer(texto)

for padrao_encontrado in padroes_encontrados:
    print(padrao_encontrado)
```

### Exemplo do prefixo de senhor e senhora

```python
texto = "Mr. Incrivel\nMr Incrivel\nMister Incrivel\nMs Simpson\nMiss Pig\nMrs. Sanders\nMrsa Vodka"

# Quero encontrar todos os nomes que comecam com Mr
# Para isso vou usar o quantificador "?" que indica um caracter opcional

# < M > exatamente isso
# < (r|s|rs) > pode ser: r OU s OU rs 
# < \.? > tem um ponto opcional
# < \s > tem um espaco
# <.*> quantos caracteres seguirem até o fim da linha
padrao = re.compile(r'M(r|s|rs)\.?\s.*')

padroes_encontrados = padrao.finditer(texto)

for padrao_encontrado in padroes_encontrados:
    print(padrao_encontrado)
```

SAÍDA:

```

<re.Match object; span=(13, 24), match='Mr Incrivel'>
<re.Match object; span=(41, 51), match='Ms Simpson'>
<re.Match object; span=(61, 73), match='Mrs. Sanders'>
```

### Exemplo filtrando nomes por DDD

```python
texto = 'Lista telefonica:\nAna - (11)1234-5678\nBeatriz - (11)4312-1223\nCamila - (19)3367-1234\nDebora - (15)2336-9972\nEloisa - (11)9982-2230'

# Dividir a lista em linhas
linhas = texto.split('\n')

# < \(11\) > Inicia com 11 entre parenteses,
# < \d{4} > tem 4 digitos,
# < \- > depois tem um traco,
# < \d{4} > depois mais 4 digitos.
padrao_telefone = re.compile(r'\(11\)\d{4}\-\d{4}') 

# <^> Começo da string, 
# <\w> caracteres de palavra,
# <+> com  de 1 ou mais, 
# <\s> até o espaço.
padrao_nome = re.compile(r'^\w+\s') 

# Analisar cada linha
for linha in linhas:
    
    # Procura se tem (11) seguido de um telefone.
    match_regiao = re.search(padrao_telefone, linha)
    
    # Se M for None -> (not M) == True
    if not match_regiao:
        continue
    else:
        # Significa que encontrou telefone com DDD (11)
        # Agora vou coletar o nome
        match_nome = re.search(padrao_nome, linha)
        if match_nome:
            comeco = match_nome.start()
            fim = match_nome.end() - 1 # O fator -1 desconsidera o espaço vazio
            print(linha[comeco:fim])
```

SAÍDA:

```
Ana
Beatriz
Eloisa
```
